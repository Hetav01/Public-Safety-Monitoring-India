## **Extract comments from YouTube json files.**

```python
import pandas as pd
import os
import json

# Define the directory containing the folders with JSON files
base_directory = "/Users/hetavpatel/Desktop/Data Science/Grad DS Work/DSCI 601 Applied Data Science/old_repos/NLPPunePorsche/PunePorscheComments/"

# List all subfolders in the base directory
folders = [f for f in os.listdir(base_directory) if os.path.isdir(os.path.join(base_directory, f))]

print(len(folders))
```

```python
# Initialize an empty list to store all comments
all_comments = []

# Define a function to process and flatten each comment
def flatten_comment(comment):
    flattened = {
        'kind': comment.get('kind'),
        'etag': comment.get('etag'),
        'id': comment.get('id'),
        'channelId': comment.get('snippet', {}).get('channelId'),
        'videoId': comment.get('snippet', {}).get('videoId'),
        'topLevelComment_kind': comment.get('snippet', {}).get('topLevelComment', {}).get('kind'),
        'topLevelComment_etag': comment.get('snippet', {}).get('topLevelComment', {}).get('etag'),
        'topLevelComment_id': comment.get('snippet', {}).get('topLevelComment', {}).get('id'),
        'textDisplay': comment.get('snippet', {}).get('topLevelComment', {}).get('snippet', {}).get('textDisplay'),
        'textOriginal': comment.get('snippet', {}).get('topLevelComment', {}).get('snippet', {}).get('textOriginal'),
        'authorDisplayName': comment.get('snippet', {}).get('topLevelComment', {}).get('snippet', {}).get('authorDisplayName'),
        'authorProfileImageUrl': comment.get('snippet', {}).get('topLevelComment', {}).get('snippet', {}).get('authorProfileImageUrl'),
        'authorChannelUrl': comment.get('snippet', {}).get('topLevelComment', {}).get('snippet', {}).get('authorChannelUrl'),
        'authorChannelId': comment.get('snippet', {}).get('topLevelComment', {}).get('snippet', {}).get('authorChannelId', {}).get('value'),
        'canRate': comment.get('snippet', {}).get('topLevelComment', {}).get('snippet', {}).get('canRate'),
        'viewerRating': comment.get('snippet', {}).get('topLevelComment', {}).get('snippet', {}).get('viewerRating'),
        'likeCount': comment.get('snippet', {}).get('topLevelComment', {}).get('snippet', {}).get('likeCount'),
        'publishedAt': comment.get('snippet', {}).get('topLevelComment', {}).get('snippet', {}).get('publishedAt'),
        'updatedAt': comment.get('snippet', {}).get('topLevelComment', {}).get('snippet', {}).get('updatedAt'),
        'canReply': comment.get('snippet', {}).get('canReply'),
        'totalReplyCount': comment.get('snippet', {}).get('totalReplyCount'),
        'isPublic': comment.get('snippet', {}).get('isPublic')
    }
    return flattened
```

```python
count = 0
# Traverse each folder and process JSON files
for folder in folders:
    folder_path = os.path.join(base_directory, folder)
    count += 1
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.json'):
            file_path = os.path.join(folder_path, file_name)
            try:
                with open(file_path, 'r', encoding='utf-8') as file:
                    data = json.load(file)
                    comments = data.get('items', [])
                    # Flatten each comment and add to the list
                    all_comments.extend([flatten_comment(comment) for comment in comments])
            except Exception as e:
                print(f"Error processing file {file_path}: {e}")
                    

# Convert all comments to a single DataFrame
df = pd.DataFrame(all_comments)
print(f"Processed {count} folders and {len(df)} comments")

# Display the consolidated DataFrame
# print(df.head())

# Save the consolidated DataFrame to a CSV file
output_csv = "/Users/hetavpatel/Desktop/Data Science/Grad DS Work/DSCI 601 Applied Data Science/old_repos/NLPPunePorsche/consolidated_youtube_comments.csv"
df.to_csv(output_csv, index=False, encoding='utf-8')
print(f"Consolidated DataFrame saved to {output_csv}")
```

```python
# Print the count of comments in each subdirectory
commentList = []
for folder in folders:
    folder_path = os.path.join(base_directory, folder)
    comment_count = 0
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.json'):
            file_path = os.path.join(folder_path, file_name)
            try:
                with open(file_path, 'r', encoding='utf-8') as file:
                    data = json.load(file)
                    comments = data.get('items', [])
                    comment_count += len(comments)
            except Exception as e:
                print(f"Error processing file {file_path}: {e}")
    commentList.append(comment_count)
    # print(f"Folder: {folder}, Comment Count: {comment_count}")
    # if (folder == "-T4KGjJ6kiw"):
    #     print(f"Folder: {folder}, Comment Count: {comment_count}")
    #     break
    
print(f"commentList: {sum(commentList)}")
print(f"Total folders processed: {count}")
```

```python
# Convert the 'publishedAt' column to datetime
df['publishedAt'] = pd.to_datetime(df['publishedAt'])
df['publishedAt']
```

```python
#find the min and max date so that we get an idea to filter the data.

print(f"Min date of a comment ever on this topic: {df["publishedAt"].min()}")
print(f"The videoId of the video with the earliest comment: {df[df["publishedAt"] == df["publishedAt"].min()]["videoId"]}")

# print("More details of the video with the earliest comment:")
# print(df[df["publishedAt"] == df["publishedAt"].min()])

print("The comment with the earliest date:")
print(df[df["publishedAt"] == df["publishedAt"].min()]["textOriginal"].iloc[0])

print("\n")

print(f"Max date of a comment ever on this topic: {df["publishedAt"].max()}")
print(f"The videoId of the video with the latest comment: {df[df["publishedAt"] == df["publishedAt"].max()]["videoId"]}")

# print("More details of the video with the latest comment:")
# print(df[df["publishedAt"] == df["publishedAt"].max()])

print("The comment with the latest date:")
print(
    df[df["publishedAt"] == df["publishedAt"].max()]["textOriginal"].iloc[0])
```

```python
# Filter the DataFrame for comments with publishedAt less than the specified date
filtered_videos = df[df['publishedAt'] < '2024-05-19 06:00:00+00:00']

# Get the unique videoIds from the filtered DataFrame
unique_video_ids = filtered_videos['videoId'].unique()

print("Videos with comments before 2024-05-19 06:00:00+00:00:")
print(unique_video_ids)
```

```python
# remove the comments from the dataframe with video ids in the unique_video_ids list

df = df[~df['videoId'].isin(unique_video_ids)]
```
