```python
import pandas as pd
df_majority = pd.read_csv('/home/sd3528/hetav-2/data/top_100_majority_human_annotate_corrected.csv')
df_labelled = pd.read_csv('/home/sd3528/hetav-2/data/2700_comments_final_labeled.csv')
```

```python
df_majority
```

```python
df_majority['human_label'] = df_majority['human_label'].apply(lambda x: x-1)
df_majority['human_label'].value_counts()
```

```python
df_labelled
```

```python
df_labelled['label'] = df_labelled['label'].apply(lambda x: x-1)
df_labelled['label'].value_counts()
```

```python
df_dict = []
for i in range(len(df_majority)):
  df_dict.append({
    'text': df_majority['text'][i],
    'label': df_majority['human_label'][i]
  })
  
for i in range(len(df_labelled)):
  df_dict.append({
    'text': df_labelled['text'][i],
    'label': df_labelled['label'][i]
  })
  
df_majority_sampling = pd.DataFrame(df_dict)
df_majority_sampling
```

```python
# train test split
df_majority_sampling = df_majority_sampling.sample(frac=1).reset_index(drop=True)
df_majority_sampling_train = df_majority_sampling[:int(len(df_majority_sampling)*0.8)]
df_majority_sampling_test = df_majority_sampling[int(len(df_majority_sampling)*0.8):]

df_majority_sampling_valid = df_majority_sampling_test[:int(len(df_majority_sampling_test)*0.5)]
df_majority_sampling_test = df_majority_sampling_test[int(len(df_majority_sampling_test)*0.5):]

len(df_majority_sampling_train), len(df_majority_sampling_test), len(df_majority_sampling_valid)
```

```python
assert (df_majority_sampling_test.iloc[0] != df_majority_sampling_valid.iloc[0]).all()
```
